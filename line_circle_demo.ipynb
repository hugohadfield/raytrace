```python
from clifford.tools.g3c import *
from pyganja import *
```

```python
C = (up(1.0*e1)^up(-1.0*e1)^up(1.0*e2)).normal()
```

```python
# First the case that the line does not pass through the circle
L1 = (up(1.5*e1)^up(1.5*e1 + 0.75*e2 + e3)^einf).normal()

S = (meet(C,L1)*I5)

gs = GanjaScene()
gs.add_objects([C], color=Color.BLACK)
gs.add_objects([L1], color=Color.RED)
gs.add_objects([S*einf*S], color=Color.BLACK, label='S')
gs.add_objects([S])
gs.add_objects([up(1.3*e2-e1)], label='S**2 = '+str((S**2)[0]))
draw(gs, scale=0.5)
```

```python
# Next the case that the line does not pass through the circle
L2 = (up(1.0*e1)^up(1.0*e1 + 0.75*e2 + e3)^einf).normal()

S2 = (meet(C,L2)*I5)

gs = GanjaScene()
gs.add_objects([C], color=Color.BLACK)
gs.add_objects([L2], color=Color.RED)
gs.add_objects([(S2*einf*S2)(1)], color=Color.BLACK, label='S')
gs.add_objects([S2])
gs.add_objects([up(1.3*e2-e1)], label='S**2 = '+str((S2**2)[0]))
draw(gs, scale=0.5)
```
